```python
ny_lib = raw_lib[raw_lib["State"] == "NY"]
ny_lib_new = ny_lib.copy()

# df is the cleaned frame to use
df = ny_lib_new.dropna() # removing all the null values 
cols = ['County Population', 'Service Population', 'Service Population Without Duplicates',
        'Librarians', 'Employees', 'Total Staff','Print Collection', 'Digital Collection', 'Audio Collection', 
        'Downloadable Audio', 'Physical Video', 'Total Licensed Databases', 'Library Visits', 'Registered Users']

# removing one row that had na values
df[cols] = df[cols].apply(pd.to_numeric,  errors='coerce',axis=1)
df.drop(df[df["County"] == '75'].index, inplace = True)
```

```python
# retrieving the number of hours employees are working weekly - total 
df["Employees_Hours"] = df["Employees"]  * 40
df = df.rename(columns = {"Employees":"Employees_NC"}) #Not Converted
```

```python
def get_dig_collection_libs(df): 
    dfc = df[df["Digital Collection"] > 0]
    return list(dfc["Library Name"])

def get_aud_collection_libs(df): 
    dfc = df[df["Audio Collection"] > 0]
    return list(dfc["Library Name"])

def get_down_collection_libs(df): 
    dfc = df[df["Downloadable Audio"] > 0]
    return list(dfc["Library Name"])

def get_pri_collection_libs(df): 
    dfc = df[df["Print Collection"] > 0]
    return list(dfc["Library Name"])

def get_pv_collection_libs(df): 
    dfc = df[df["Physical Video"] > 0]
    return list(dfc["Library Name"])
```

```python
# DIGITAL COLLECTION 

# get all the digital collection libraries
dc = df.groupby("County").apply(get_dig_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have digital collection 
df = df.merge(dc, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "digital_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Digital Collection"] == 0, "digital_collection_links"] = 0

# # removing self from list of links 
# for row in df.itertuples():
#     if row.digital_collection_links != 0:
#         library = row._2
#         digital_collection_links = list(row.digital_collection_links)
#         digital_collection_links.remove(library)
#         df.at[row.Index, 'digital_collection_links'] = digital_collection_links
```

# creating json for geometric points 

```python
dict_location = {}
df_loc = df.copy()
suffolk_loc = df_loc[df_loc["County"] == "SUFFOLK"]
suffolk_loc = suffolk_loc[["Library Name", "Longitude", "Latitude", "Librarians", "Employees_Hours"]]
points_suffolk = {"type":"FeatureCollection","features":[]}

count = 1
for row in suffolk_loc.itertuples():
    loc_info = {}
    library_name = row._1
    employee_hours = row.Employees_Hours
    librarians = row.Librarians
    loc_info["type"] = "Feature"
    if count < 10: 
        loc_info["id"] = "0"+str(count)
    else: 
        loc_info["id"] = str(count)
    loc_info["properties"] = {"name":library_name, "employee_hours": employee_hours, "librarians":librarians}
    loc_info["geometry"] = {
        "type":"Point", 
        "coordinates":[row.Longitude, row.Latitude]
    }
    points_suffolk["features"].append(loc_info)
    
dict_location["SUFFOLK"] = points_suffolk
```

```python
nassau_loc = df_loc[df_loc["County"] == "NASSAU"]
nassau_loc = nassau_loc[["Library Name", "Longitude", "Latitude", "Librarians", "Employees_Hours"]]
points_nassau = {"type":"FeatureCollection","features":[]}

count = 1
for row in nassau_loc.itertuples():
    loc_info = {}
    library_name = row._1
    employee_hours = row.Employees_Hours
    librarians = row.Librarians
    loc_info["type"] = "Feature"
    if count < 10: 
        loc_info["id"] = "0"+str(count)
    else: 
        loc_info["id"] = str(count)
    loc_info["properties"] = {"name":library_name, "employee_hours": employee_hours, "librarians":librarians}
    loc_info["geometry"] = {
        "type":"Point", 
        "coordinates":[row.Longitude, row.Latitude]
    }
    points_nassau["features"].append(loc_info)

dict_location["NASSAU"] = points_nassau
```

```python
westchester_loc = df_loc[df_loc["County"] == "WESTCHESTER"]
westchester_loc = westchester_loc[["Library Name", "Longitude", "Latitude", "Librarians", "Employees_Hours"]]
points_westchester = {"type":"FeatureCollection","features":[]}

count = 1
for row in westchester_loc.itertuples():
    loc_info = {}
    library_name = row._1
    employee_hours = row.Employees_Hours
    librarians = row.Librarians
    loc_info["type"] = "Feature"
    if count < 10: 
        loc_info["id"] = "0"+str(count)
    else: 
        loc_info["id"] = str(count)
    loc_info["properties"] = {"name":library_name, "employee_hours": employee_hours, "librarians":librarians}
    loc_info["geometry"] = {
        "type":"Point", 
        "coordinates":[row.Longitude, row.Latitude]
    }
    points_westchester["features"].append(loc_info)

dict_location["WESTCHESTER"] = points_westchester
```

# create json for node-links 

```python
dict_three_states = {}
```

```python
suffolk = df[df["County"] == "SUFFOLK"]
suffolk = suffolk[["Library Name", 
         "Service Population", 
         "Print Collection", 
         "print_collection_links"
        ]] 

data_suffolk = {}
data_suffolk["nodes"] = []
data_suffolk["links"] = []

count_nodes = 1
track_info = {}
track_value = {}
# adding in the nodes right now
for row in suffolk.itertuples():
    library_name = row._1
    info = {"id":count_nodes, "name":library_name}
    data_suffolk["nodes"].append(info)
    track_info[library_name] = count_nodes
    track_value[library_name] = row._3
    count_nodes += 1
    
for row in suffolk.itertuples():
    library_name = row._1
    links = row.print_collection_links
    links.remove(library_name)
    for i in links: 
        source = track_info[library_name]
        target = track_info[i]
        value = track_value[library_name]
        info_links = {"source":source, "target":target, "value":value}
        data_suffolk["links"].append(info_links)
        
dict_three_states["SUFFOLK"] = data_suffolk
```

```python
nassau = df[df["County"] == "NASSAU"]
nassau = nassau[["Library Name", 
         "Service Population", 
         "Print Collection", 
         "print_collection_links"
        ]]

data_nassau = {}
data_nassau["nodes"] = []
data_nassau["links"] = []

count_nodes = 1
track_info = {}
track_value = {}
# adding in the nodes right now
for row in nassau.itertuples():
    library_name = row._1
    info = {"id":count_nodes, "name":library_name}
    data_nassau["nodes"].append(info)
    track_info[library_name] = count_nodes
    track_value[library_name] = row._3
    count_nodes += 1
    
for row in nassau.itertuples():
    library_name = row._1
    links = row.print_collection_links
    links.remove(library_name)
    for i in links: 
        source = track_info[library_name]
        target = track_info[i]
        value = track_value[library_name]
        info_links = {"source":source, "target":target, "value":value}
        data_nassau["links"].append(info_links)
        
dict_three_states["NASSAU"] = data_nassau
```

```python
westchester = df[df["County"] == "WESTCHESTER"]
westchester = westchester[["Library Name", 
         "Service Population", 
         "Print Collection", 
         "print_collection_links"
        ]] 

data_westchester = {}
data_westchester["nodes"] = []
data_westchester["links"] = []

count_nodes = 1
track_info = {}
track_value = {}
# adding in the nodes right now
for row in westchester.itertuples():
    library_name = row._1
    info = {"id":count_nodes, "name":library_name}
    data_westchester["nodes"].append(info)
    track_info[library_name] = count_nodes
    track_value[library_name] = row._3
    count_nodes += 1

# adding in all the links 
for row in westchester.itertuples():
    library_name = row._1
    links = row.print_collection_links
    links.remove(library_name)
    for i in links: 
        source = track_info[library_name]
        target = track_info[i]
        value = track_value[library_name]
        info_links = {"source":source, "target":target, "value":value}
        data_westchester["links"].append(info_links)
        
dict_three_states["WESTCHESTER"] = data_westchester
```
