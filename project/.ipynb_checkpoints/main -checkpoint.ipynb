```python
ny_lib = raw_lib[raw_lib["State"] == "NY"]
ny_lib_new = ny_lib.copy()

# df is the cleaned frame to use
df = ny_lib_new.dropna() # removing all the null values 
cols = ['County Population', 'Service Population', 'Service Population Without Duplicates',
        'Librarians', 'Employees', 'Total Staff','Print Collection', 'Digital Collection', 'Audio Collection', 
        'Downloadable Audio', 'Physical Video', 'Total Licensed Databases', 'Library Visits', 'Registered Users']

# removing one row that had na values
df[cols] = df[cols].apply(pd.to_numeric,  errors='coerce',axis=1)
df.drop(df[df["County"] == '75'].index, inplace = True)
```

```python
df
```

```python
# retrieving the number of hours employees are working weekly - total 
df["Employees_Hours"] = df["Employees"]  * 40
df = df.rename(columns = {"Employees":"Employees_NC"}) #Not Converted
```

```python
def get_dig_collection_libs(df): 
    dfc = df[df["Digital Collection"] > 0]
    return list(dfc["Library Name"])

def get_aud_collection_libs(df): 
    dfc = df[df["Audio Collection"] > 0]
    return list(dfc["Library Name"])

def get_down_collection_libs(df): 
    dfc = df[df["Downloadable Audio"] > 0]
    return list(dfc["Library Name"])

def get_pri_collection_libs(df): 
    dfc = df[df["Print Collection"] > 0]
    return list(dfc["Library Name"])

def get_pv_collection_libs(df): 
    dfc = df[df["Physical Video"] > 0]
    return list(dfc["Library Name"])
```

```python
# DIGITAL COLLECTION 

# get all the digital collection libraries
dc = df.groupby("County").apply(get_dig_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have digital collection 
df = df.merge(dc, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "digital_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Digital Collection"] == 0, "digital_collection_links"] = 0

# removing self from list of links 
for row in df.itertuples():
    if row.digital_collection_links != 0:
        library = row._3
        digital_collection_links = list(row.digital_collection_links)
        digital_collection_links.remove(library)
        df.at[row.Index, 'digital_collection_links'] = digital_collection_links
```

```python
# AUDIO COLLECTION 

# get all the audio collection libraries
ac = df.groupby("County").apply(get_aud_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have audio collection 
df = df.merge(ac, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "audio_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Audio Collection"] == 0, "audio_collection_links"] = 0

# removing self from list of links 
for row in df.itertuples():
    if row.audio_collection_links != 0:
        library = row._3
        audio_collection_links = list(row.audio_collection_links)
        audio_collection_links.remove(library)
        df.at[row.Index, 'audio_collection_links'] = audio_collection_links
```

```python
# PRINT COLLECTION 

# get all the audio collection libraries
pc = df.groupby("County").apply(get_pri_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have print collection 
df = df.merge(pc, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "print_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Print Collection"] == 0, "print_collection_links"] = 0

# removing self from list of links 
for row in df.itertuples():
    if row.print_collection_links != 0:
        library = row._3
        print_collection_links = list(row.print_collection_links)
        print_collection_links.remove(library)
        df.at[row.Index, 'print_collection_links'] = print_collection_links
```

```python
# DOWNLOADABLE COLLECTION 

# get all the audio collection libraries
downc = df.groupby("County").apply(get_down_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have print collection 
df = df.merge(downc, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "download_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Downloadable Audio"] == 0, "download_collection_links"] = 0

# removing self from list of links 
for row in df.itertuples():
    if row.download_collection_links != 0:
        library = row._3
        download_collection_links = list(row.download_collection_links)
        download_collection_links.remove(library)
        df.at[row.Index, 'download_collection_links'] = download_collection_links
```

```python
# DOWNLOADABLE COLLECTION 

# get all the audio collection libraries
vc = df.groupby("County").apply(get_pv_collection_libs).to_frame().reset_index()
# add in all the libraries within a county that have print collection 
df = df.merge(vc, on = "County")

# rename columns so it has dc links
df.rename(columns = {0: "video_collection_links"}, inplace = True)
# replace any row that has 0 for a colleciton with 0 digital links
df.loc[df["Physical Video"] == 0, "video_collection_links"] = 0

# removing self from list of links 
for row in df.itertuples():
    if row.video_collection_links != 0:
        library = row._3
        video_collection_links = list(row.video_collection_links)
        video_collection_links.remove(library)
        df.at[row.Index, 'video_collection_links'] = video_collection_links
```
